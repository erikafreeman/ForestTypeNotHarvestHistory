```python
import numpy as np
#exponential

def addStepDecayToDF(df):
    #binary step
    for T in np.linspace(1,99,99): #lambda parameter stepping through 400
        new_col = []
        index = 0
        for i in Year_to_Pr:
            if i > T:
                weighted_area = 0 * (df["area_harve"][index])
               # print(weighted_area)
                new_col.append(weighted_area)
                index +=1
            
            else:
                weighted_area =  df["area_harve"][index]
                new_col.append(weighted_area)
                index +=1
        df['stp_'+ str(T)[0:4]]= new_col
    # return df

def addExpToDF(df):
    #exponential
    for lmnda in np.linspace(0.001,0.9,50): #lambda parameter stepping through 400
        lamb = lmnda
        decay_param = np.linspace(1,99,99) #decay paramter should step 1-100
        weight = np.exp(-lamb * decay_param)
        mlt.plot(weight)
        new_col = []
        for i in Year_to_Pr:
            index = 0
            new_weight = np.interp(i, decay_param, weight)
            weighted_area = new_weight * df["area_harve"][index]
            new_col.append(weighted_area)
            index +=1
        df['ex_'+ str(lamb)[0:4]]= new_col
            
def addLinearDecayToDF(df):
    for T in np.linspace(1,99,99):
      new_col = []
      i=0
      minWeight = 100000
      maxWeight = -1000
      for area, ytp in zip(df["area_harve"], Year_to_Pr):
        weight = max(0,(1-ytp/T))
        weightedArea = weight*area
        new_col.append(weightedArea)
        i += 1
        minWeight = min(weight, minWeight)
        maxWeight = max(weight, maxWeight)

      # print('minWeight:', minWeight)
      # print('maxWeight:', maxWeight)
      df['lin_'+ str(T)[0:4]]= new_col

def addInversePowerToDF(df):
    for beta in np.linspace(.5,1.5,100):
      new_col = []
      i=0
      allWeights = []
      # Year_to_Pr = list(Year_to_Pr)
      # Year_to_Pr.sort()
      # minWeight = 100000
      # maxWeight = -1000
      for area, ytp in zip(df["area_harve"], Year_to_Pr):
        weight = np.power(ytp, -beta)
        weightedArea = weight*area
        allWeights.append(weight)
        # print('area:', area)
        # print('weightedArea:', weightedArea)
        new_col.append(weightedArea)
        i += 1
        # minWeight = min(weight, minWeight)
        # maxWeight = max(weight, maxWeight)
      
      # print('minWeight:', minWeight)
      # print('maxWeight:', maxWeight)
      # mlt.plot(allWeights)
      # mlt.clear()
      df['invp_'+ str(beta)[0:4]]= new_col

addStepDecayToDF(Harvest)
addExpToDF(Harvest)
addLinearDecayToDF(Harvest)
addInversePowerToDF(Harvest)
# print(list(Harvest.columns.values))

display(Harvest)

# Harvest1= linear_decay(Harvest)
# Harvest2= exp_decay(Harvest)
# Harvest3 = step_decay(Harvest2)
# display(list(Harvest3.columns))
#weight_list = pd.DataFrame(
#    {'lambda': lb,
#     'decay_param': dp, #match for each lambda to the decay param
#     'weight': wt
#    })

#display(weight_list)
```
